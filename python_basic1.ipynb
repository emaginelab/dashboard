## 튜플(Tuple)  
- 리스트 자료와 유사, 읽기 전용 자료구조
- 튜플은 값을 변경할 수 없음
- 튜플 표현 : 괄호(),  (예) num=(1, 3, 2, 4)
- 튜플에 한 개 값 표현은 값 뒤에 반드시 콤마 작성, (예) num=(10,)
- 연속으로 나열된 값을 저장하면 튜플 자료형이 됨 (예) num=1, 3, 2, 4
- 튜플의 인덱스, 인덱싱, 슬라이싱은 리스트와 동일
- 패킹과 언패킹   
```
num = 3, 1, 7  #패킹
num1, num2, num3 = num  #언패킹
```

##딕셔너리(Dictionary)
- 키와 값을 쌍으로 저장하는 자료구조
- 딕셔너리 표현 : 중괄호 {}, {키:값}
- 딕셔너리 값 추가 : **딕셔너리[키] = 값** #키가 딕셔너리에 없는 경우
- 딕셔너리 값 변경 : **딕셔너리[키] = 값** #키가 딕셔너리에 있는 경우
- 딕셔너리 값 삭제 : **del(딕셔너리[키])**
- 딕셔너리 값 검색 : **딕셔너리[키]**, **딕셔너리.get(키)**
- 딕셔너리의 모든 키만 리스트로 반환 : **딕셔너리.keys()**
- 딕셔너리의 모든 값만 리스트로 반환 : **딕셔너리.values()**
- 딕셔너리의 모든 키와 값을 리스트로 반환 : **딕셔너리.items()**  

```
dt = {'a':1, 'b':2, 'c':3}
dt['d'] = 4
dt['c'] = 30
dt[2]
dt.keys()
dt.values()
dt.items()
```

## if 조건문
- 관계연산자(조건연산자)
  > 두 값을 비교하기 위한 연산자  
  > 선택처리, 반복처리에서 사용  
  > 결과는 항상 참(True), 거짓(False) 반환  

- 조건식  
  > 나이(age)가 18세 미만인가?  # age < 18  
  > 두 사람의 나이(age1, age2) 합이 100인가?  # age1 + age2 = 100

1. 단순 if문(if ~)  
  > 조건식 결과가 참인 경우에만 지정 문장 수행  
  ```
  if 조건식 : 
       문장
  ```

````python
score = 80
if score > 80 :
  print('합격')
````

2. 이중 if문(if ~ else ~)  
  > 조건식의 참, 거짓 결과에 따라 서로 다른 작업을 수행  
  ```
  if 조건식 : 
        문장1
  else : 
        문장2
  ```

````python
score = 80
if score > 80 :
  print('합격')
else :
  print('불합격')
````

3. 다중 if문(if ~ elif ~ else ~)  
  > 조건식의 결과에 따라 각각 처리하는 조건문  
  ```
  if 조건식1 : 
        문장1
  elif 조건식2 : 
        문장2
        :
  else : 
        문장3
  ```

````python
score = 80
if score >= 80 :
  print('A학점')
elif score >= 60 :
  print('B학점')
else :
  print('C학점')
````

4. 논리연산자  
- not : 조건식의 결과 논리값를 반대로 변경
- and : 조건식의 결과가 모두 참인 경우 참 명령 실행
- or : 조건식의 결과가 한 개라도 참인 경우 참 명령 실행 

조건식1 | 조건식2 | and 논리값
---|---|---
False|Fase|False
False|True|False
True|False|False
True|True|True  

조건식1 | 조건식2 | or 논리값
---|---|---
False|Fase|False
False|True|True
True|False|True
True|True|True

````python
kor, eng = 90, 90
# and
if kor>=80 and eng>=80 :
  print('합격')
else :
  print('불합격')
# or
if kor>=80 or eng>=80 :
  print('합격')
else :
  print('불합격')
````
