- 문자형의 연산 : __+__(문자열 연결), __*__(문자열 반복)

````python
print("안녕" + "파이썬")
````

````python
print("안녕" * 3)
````

- 문자형 인덱스와 인덱싱
 > 인덱스 : 위치(주소), 인덱스는 0부터 시작, 역인덱스는 -1부터 시작  
 > 인덱싱 : 자료 접근 / 대괄호로 표현 : 문자열[위치]

````python
print("안녕파이썬"[0])
````

````python
print("안녕파이썬"[-2])
````

 - 문자형 슬라이싱
 > 슬라이싱 : **[시작위치:끝위치-1]**  
    - 시작위치 생략:0부터  
    - 끝위치 생략:끝까지

````python
print("안녕파이썬프로그래밍"[0:3])
print("안녕파이썬프로그래밍"[:4])
print("안녕파이썬프로그래밍"[1:])
print("안녕파이썬프로그래밍"[3:7])
````

- 문자열 길이 구하는 함수 : __len()__

````python
print(len("안녕파이썬"))
````

##변수와 연산처리

- 변수란 ?  
>데이터를 저장하기 위한 기억 장소  

- 변수명 작성 규칙  
> 영문자, 숫자, 밑줄문자(_) 등이 조합으로 작성, 한글도 사용 가능  
> 첫 글자는 영문자 또는 밑줄문자만 가능  
> 공백을 포함할 수 없음  
> 대소문자 서로 구별  
> 예약어(명령어)는 변수명으로 사용 못함

- 대입 연산 처리  
> 변수 = 값(또는 변수, 수식)  

  ```
str_var = "파이썬"
num_var = 100
tot = 10 + 20
result = tot  
  ```
- 여러 개 값 대입하기
> 변수, 변수, 변수 ...  = 값, 값, 값 .... (변수와 값을 1:1 대응)  

  ```
num1, num2, num3 = 10, 20, 30
  ```

````python
str_var = "파이썬"
print(str_var)
num_var = 100
print(num_var)
tot = 10 + 20
print(tot)
result = tot
print(result)  
````

````python
num1, num2, num3 = 10, 20, 30
````
